```python
import pandas as pd
from googletrans import Translator

df = pd.read_csv('../markettwits.tsv', sep='\t')
df['Text'] = df['Text'].str.replace('\\n', '\n').str.replace('\\t', '\t')

bs = 100
i = 0
n = len(df)
translations = []

df
```

```python
async with Translator() as translator:
    while i <= n:
        results = await translator.translate(list(df['Text'][i:i+bs]), src='ru')
        translations.extend(res.text for res in results)
        i += bs
```

```python
out_df = df.copy()[:len(translations)]
out_df['Text'] = translations
out_df['Text'] = out_df['Text'].str.replace('\n', '\\n').str.replace('\t', '\\t')
out_df
```

```python
out_df.to_csv('eng_markettwits.tsv', index=False, sep='\t')
```

```python
import pandas as pd
from googletrans import Translator

df = pd.read_csv('../markettwits.tsv', sep='\t')
df = df[df['Date'] > '2025-02-01 00:00:01+00:00']

async def translate_list(list):
    results = []
    async with Translator() as translator:
        for s in list:
            result = await translator.translate(str(s).replace('\\n', '\n').replace('\\t', '\t'), src='ru', dest='en')
            results.append(result.text.replace('\n', '\\n').replace('\t', '\\t'))
    return results

translated_texts = await translate_list(df['Text'].to_list())
list_df = pd.DataFrame({'EnglishText' : translated_texts})
merged_df = pd.concat([df, list_df], axis=1)
merged_df.to_csv('translated_output.tsv', index=False)
```
